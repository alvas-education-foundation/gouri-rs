print multiples of 4 upto 150 using a list

```
a=[]
c=1
for i in range(0,150):
    m=4*c
    if m<200:
        a.append(m)
        c=c+1
print(a,end="")
```

print the nos only divisable by 4 and 5 between 100 and 200 using a list

```
a=[]
b=[]
for i in range(100,200):
    if i%4==0:
        a.append(i)
    if i%5==0:
        b.append(i)
print("The numbers which are divisible by 4 are:",a,end="")
print("\n")
print("The numbers which are divisible by 5 are:",b,end="")
```

Add the n number of names in a list and print them alphabetically and reverse alphabetically

```
my_str = input("Enter a string: ")

# breakdown the string into a list of words
words = my_str.split()

# sort the list
words.sort()

# display the sorted words

print("The sorted words are:")
for word in words:
   print(word)

#reverse the words in list
words.reverse()

#display the reversed words
print("The reverse words are:")
for wor in words:
   print(wor)
```

Print perfect squares and divisible by 5 between 200 and 400( (both inclusive) using list

```
a=[]
b=[]
c=[]
import math
for i in range(200,400):
    sr = math.sqrt(i)
    if sr-math.floor(sr)==0:
        a.append(i)
        b.append(sr)
    if i%5==0:
        c.append(i)
print("Perfect squares in the given range are:",a,end=" ")
print("\n")
print("Squares of above numbers are:",b,end=" ")
print("\n")
print("Numbers which are divisible by 5 are:",c,end=" ")
```

Write python program to get a list of tuples of Rollno,Name for 5 students through keyboard and sort them Rollno wise descending order

```
def Sort_Tuple(tup):  
    lst = len(tup)  
    for i in range(0, lst):  
        for j in range(0, lst-i-1):  
            if (tup[j][1]<tup[j + 1][1]):  
                temp = tup[j]  
                tup[j]= tup[j + 1]  
                tup[j + 1]= temp 
    return tup
tup =[]
num = int(input("Number of students:"))
for i in range(0,num):
     name = input(" Name of student: ")
     rollnum = input("Roll num of the student:")
     tup.append((name,rollnum))
print(Sort_Tuple(tup))
```
